```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import matplotlib.patches as patches
import scipy.io

#read, set figure, show img
ref_img = plt.imread('img_000.png')
fig, ax = plt.subplots(figsize=[25,15])
plt.imshow(ref_img[510:1379, 300:1048], cmap='gray')

#patch formatting
linewidth = 3
edgecolor = 'y'
ls1 = '--'
ls2 = ':'
facecolor = 'none'

#text formatting
text_c = 'b'
font_s = 12
font_w = 1000
text_box = dict(boxstyle='round',
                   ec=(1., 0.5, 0.5),
                   fc=(1., 0.8, 0.8),
                   )

#patch region 1
rect1 = patches.Rectangle((330, 385), 125, 130,
                          linewidth=linewidth, edgecolor=edgecolor, ls=ls1, facecolor=facecolor
                         )
ax.add_patch(rect1)
plt.text(340, 425, 'Volume 1\n1 element',
         color = text_c, fontsize = font_s, fontweight = font_w, bbox=text_box
        )

#patch region 2
rect2 = patches.Rectangle((265, 320), 255, 255,
                          linewidth=linewidth, edgecolor=edgecolor, ls=ls1, facecolor=facecolor
                         )
ax.add_patch(rect2)
plt.text(275, 355, 'Volume 2\n4 element',
         color = text_c, fontsize = font_s, fontweight = font_w, bbox=text_box
        )

#patch region 3
rect3 = patches.Rectangle((200, 255), 380, 385,
                          linewidth=linewidth, edgecolor=edgecolor, ls=ls1, facecolor=facecolor
                         )
ax.add_patch(rect3)
plt.text(220, 290, 'Volume 3\n9 element',
         color = text_c, fontsize = font_s, fontweight = font_w, bbox=text_box
        )

#patch region 4
rect4 = patches.Rectangle((134, 193), 507, 507, 
                          linewidth=linewidth, edgecolor=edgecolor, ls=ls1, facecolor=facecolor
                         )
ax.add_patch(rect4)
plt.text(154, 228, 'Volume 4\n16 element',
         color = text_c, fontsize = font_s, fontweight = font_w, bbox=text_box
        )

#patch region 5
rect5 = patches.Rectangle((80, 125), 615, 640,
                          linewidth=linewidth, edgecolor=edgecolor, ls=ls1, facecolor=facecolor
                         )
ax.add_patch(rect5)
plt.text(100, 160, 'Volume 5\n25 element',
         color = text_c, fontsize = font_s, fontweight = font_w, bbox=text_box
        )
```
