```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io

# baca data temp
data1 = scipy.io.loadmat('Probe4_Thermo.mat')
temp_data = data1['series_image_kelvin']
temp_dim = np.array(irt_data.shape)
temp_img1 = temp_data[150:790,110:720,0]
temp_img300 = temp_data[150:790,110:720,335]

# setup subplot
fig, (ax1,ax2) = plt.subplots(nrows=1,ncols=2,figsize=[20,8])

# plot info
ax1.set_title('Temperatur (Kelvin), IRT data baru step 000')
ax2.set_title('Temperatur (Kelvin), IRT data baru step 335')

# plot
im1 = ax1.imshow(temp_img1, vmin=284.25, vmax=285.1, cmap = 'jet')
fig.colorbar(im1, ax=ax1, shrink=1)

im2 = ax2.imshow(temp_img300, vmin=284.25, vmax=285.1, cmap = 'jet')
fig.colorbar(im2, ax=ax2, shrink=1)

# slice
ax1.axvline(x=90,ymin=0,ymax=1,color='red')
ax2.axvline(x=90,ymin=0,ymax=1,color='red')
```

```python
#slice data
temp_slice1 = temp_img1[:,95]
x_val1 = np.linspace(0,len(temp_slice1),len(temp_slice1))

temp_slice300 = temp_img300[:,95]
x_val300 = np.linspace(0,len(temp_slice300),len(temp_slice300))

#setup fig
plt.figure(figsize=[15,6])
plt.title('Slice (garis merah) (Kelvin)', fontdict={
    'fontsize': 16,
})
plt.grid()

#plot
plt.plot(x_val1,temp_slice1, label = 'step 000')
plt.plot(x_val300,temp_slice300, label = 'step 300')
plt.legend()
plt.show()
```

```python
# baca data strain
data2 = scipy.io.loadmat('eyy_cur_step_335.mat')
eyy_335 = data2['eyy_cur'][0][0]
eyy_img = eyy_335[170:460, 100:350]

data3 = scipy.io.loadmat('exx_cur_step_335.mat')
exx_335 = data3['exx_cur'][0][0]
exx_img = exx_335[170:460, 100:350]

# setup subplot
fig, (ax1,ax2) = plt.subplots(nrows=1,ncols=2,figsize=[20,8])

# plot info
ax1.set_title('eyy step 335')
ax2.set_title('exx step 335')

# plot
im1 = ax1.imshow(eyy_img, vmin = -0.14, vmax = 0.14, cmap='bwr')
fig.colorbar(im1, ax=ax1, shrink=1)

im2 = ax2.imshow(exx_img, vmin = -0.1, vmax = 0.1, cmap='bwr')
fig.colorbar(im2, ax=ax2, shrink=1)

print('max eyy',"=",np.ndarray.max(eyy_img))
print('min eyy',"=",np.ndarray.min(eyy_img))

print('max exx',"=",np.ndarray.max(exx_img))
print('min exx',"=",np.ndarray.min(exx_img))

# slice
ax1.axvline(x=54,ymin=0,ymax=1,color='red')
ax2.axvline(x=63,ymin=0,ymax=1,color='red')
```

```python
#slice data
eyy_slice = eyy_img[:,54]
x_val_eyy = np.linspace(0,len(eyy_slice),len(eyy_slice))

exx_slice = exx_img[:,63]
x_val_exx = np.linspace(0,len(exx_slice),len(exx_slice))

#setup fig
plt.figure(figsize=[15,6])
plt.title('Slice (garis merah) (Kelvin)', fontdict={
    'fontsize': 16,
})
plt.grid()

#plot
plt.plot(x_val_eyy,eyy_slice, label = 'eyy')
plt.plot(x_val_exx,exx_slice, label = 'exx')
plt.legend()
plt.show()
```
